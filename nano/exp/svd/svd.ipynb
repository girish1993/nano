```python
%matplotlib inline
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

```python
# read the image
img_np = np.array(Image.open("/home/gbhatta/personal/nano/nano/assets/sample.jpg"))
plt.imshow(img_np, cmap='gray')
```

```python
# applying numpy svd to get factors

U, S, V_T = np.linalg.svd(img_np)
print(" U shape", U.shape)
print(" S shape", S.shape)
print(" V_T shape", V_T.shape)
```

```python
# getting the top K singular values
def get_top_k_singluar_values(k: int):
    S_k = np.diag(S[:k])
    U_k = U[:, 0:k]
    V_T_k = V_T[0:k, :]
    print(" U_k shape", U_k.shape)
    print(" S_k shape", S_k.shape)
    print(" V_T_k shape", V_T_k.shape)
    return U_k, S_k, V_T_k
```

```python
# top 10 singluar values
k = 10
U_k, S_k, V_T_k = get_top_k_singluar_values(k)
compressed_img = U_k @ S_k @ V_T_k
plt.imshow(compressed_img, cmap='gray')
```

```python
# top 50 singluar values
k = 50
U_k, S_k, V_T_k = get_top_k_singluar_values(k)
compressed_img = U_k @ S_k @ V_T_k
plt.imshow(compressed_img, cmap='gray')
```
